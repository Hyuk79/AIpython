```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df_iris = pd.read_csv('Iris.csv')
```

```python
print(len(df_iris))
print(len(df_iris.columns))
```

```python
print(df_iris.info())
print(df_iris.isnull().sum())
print(df_iris.isna().sum())
```

```python
Sepal_mean = df_iris['SepalLengthCm'].mean()
Sepal_mean
```

```python
Petal_max = df_iris['PetalWidthCm'].max()
Petal_max
```

```python
species_iris = df_iris['Species'].value_counts()
species_iris
```

```python
petal_length = df_iris[df_iris['PetalLengthCm']>2.0]
len(petal_length)
```

```python
sepal_width = df_iris[df_iris['SepalWidthCm']>3.5]
len(sepal_width)
```

```python
versicolor_df = df_iris[df_iris['Species']=='Iris-versicolor']
data1 = versicolor_df['PetalLengthCm'].mean()
data1
```

```python
virginica_df = df_iris[df_iris['Species']=='Iris-virginica']
data2 = virginica_df['SepalWidthCm'].max()
data2
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df_iris,x='PetalLengthCm',kde=True, bins=15)
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_iris,x='PetalLengthCm',y='PetalWidthCm')
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=df_iris,x='PetalLengthCm',y='PetalWidthCm',hue='Species',style='Species', s=100)
plt.grid(True,linestyle='--',alpha=0.7)
plt.show()
```

```python
result1 = df_iris.groupby('Species')[['SepalLengthCm', 'PetalWidthCm']].mean()
result1
```

```python
result = df_iris.groupby('Species')['PetalLengthCm'].mean(numeric_only=True)
plt.figure(figsize = (8,6))
result.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,12))
fig.suptitle('',fontsize=16)
sns.boxplot(ax=axes[0,0],data=df_iris,x='Species',y='SepalLengthCm')
sns.boxplot(ax=axes[0,1],data=df_iris,x='Species',y='SepalWidthCm')
sns.boxplot(ax=axes[1,0],data=df_iris,x='Species',y='PetalLengthCm')
sns.boxplot(ax=axes[1,1],data=df_iris,x='Species',y='PetalWidthCm')
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```
